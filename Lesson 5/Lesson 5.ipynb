```
import random
list_with_numbers = []

for i in range(10):
    list_with_numbers.append(random.choice(range(100)))

list_with_numbers
```

```
for n in list_with_numbers:
    if n % 2 == 0:
        print(n)
```

```
for i in range(len(list_with_numbers)):
    if list_with_numbers[i] % 2 != 0:
        print(i, list_with_numbers[i])
```

```
for i, v in enumerate(list_with_numbers):
    if v % 2 != 0:
        print(i, v)
```
