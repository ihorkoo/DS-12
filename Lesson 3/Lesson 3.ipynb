```
# якщо число ділиться на 2 і на 3, то вивести А
# якщо число ділиться тільки на 2, то вивести B
# якщо число ділиться тільки на 3, то вивести C

number = int(input('Введіть число: '))

# (number % 2) and (number % 3)
if number %  6 == 0:
    print('A')
elif number %  2 == 0:
    print('B')
elif number %  3 == 0:
    print('C')
```
