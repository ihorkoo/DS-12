```
import time


def timer(func):
    def wrapper():
        start = time.time()
        res = func()
        stop = time.time()
        print(stop-start)
        return res
    return wrapper



@timer
def foo():
    [i for i in range(9999999)]
    return 

@timer
def goo():
    123 + 134
    return 

# start = time.time()
# foo()
# stop = time.time()

# print(stop - start)
```

Дано два цілих числа N, K
	
Створіть текстовий файл. Запишіть у нього N рядків, кожна із яких складається із K символів "*" (зірочка)

```
N = 3
K = 5

f = open('my_file.txt', 'w')

new_list = ['*'*K+'\n' for i in range(N)]

f.writelines(new_list)

f.close()
```
